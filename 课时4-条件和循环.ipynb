>## 1.条件知识点
[返回目录](#目录)

Python 条件语句是通过一条或多条语句的执行结果（True或者False）来决定执行的代码块。  
既然是代码块，就需要注意两点，一是缩进，Python是用缩进来标识出哪一段属于本循环，二是条件后面有冒号：

Python 编程中 if 语句用于控制程序的执行，基本形式为：

    if 判断条件：  
        执行语句……  
    else：  
        执行语句……  

下图来简单描述了条件语句的执行过程:


Python程序语言指定任何非0和非空（null）值为true，0 或者 null为false。

### 1.1IF单条件语句

```python
# 某个股票的 5日和 20日均线，判断信号
stockma5 = 19.34
stockma20 = 18.45
```

```python
# 语句
if stockma5 > stockma20:
    print('5日大于20日均线，买入')
else:
    print('5日小于20日均线，卖出')
```

if 语句的判断条件可以用>（大于）、<(小于)、==（等于）、>=（大于等于）、<=（小于等于）来表示其关系。

### 1.2IF多条件语句

如果判断需要多个条件需同时判断时：
+ 使用 or（ 或）时，表示只要两个条件有一个成立时，判断条件正确
+ 使用 and（且）时，表示只有两个条件同时成立的情况下，判断条件才证券

```python
# 某个股票的5日和20日均线，和MACD值，判断双金叉
stockma5 = 13.34
stockma20 = 20.45
stockmacd = 0.4
```

```python
# 且
if (stockma5 > stockma20) and (stockmacd > 0):
    print('5日大于20日均线,或者 MACD大于0，买入')
else:
    print('None')
```

```python
# 或
if (stockma5 > stockma20) or (stockmacd > 0):
    print('5日大于20日均线 或 MACD大于0，买入')
else:
    print('None')
```

需注意：
+ 当if有多个条件时可使用括号来区分判断的先后顺序，括号中的判断优先执行
+ \>（大于）、<（小于）等判断符号的优先级要高于 and 和 or 的

### 1.3多条件多语句

对于多条件，else要写成elif，标准格式为：

```python
if price == lowlimit:
    print('当前已跌停')
elif price == highlimit:
    print('当前已涨停')
else:
    print('当前未涨跌停')
```

注意 if... elif... else...三个是并列的，不能有缩进：

```python
'''
使用if条件，编成程序：判断某只股票RSI指标强弱。
'''

# 输入某只股票的RSI值【0-100】
stockRSI = 20

# 判断<20
if stockRSI<=20:
    # 输出
    print('该股票目前弱，不建议买入')
# 再次判断 20-50
elif stockRSI > 20 and stockRSI < 50:
    # 输出
    print('该股票目前较弱，建议观望')
# 再次判断
elif stockRSI == 50:
    # 输出
    print('该股票目前方向不明确，建议观望')
# 再次50-70
elif stockRSI > 50 and stockRSI < 70:
    #输出
    print('该股票目前较好，建议买一点')
# 再次判断 >70
elif stockRSI>=70:
    # 输出
    print('该股票目前好，建议买入')
```

循环语句允许我们执行一个语句或语句组多次，下面是在大多数编程语言中的循环语句的一般形式：

### 2.1while循环

#### 2.1.1while语句执行过程

程序执行到 while 处，“当”条件为 True 时，就去执行 while 内部的代码；“当”条件为 False 时，就跳过。   
语法为：

    while 条件:
        循环执行的语句

 while 语句用于循环执行程序，即在某条件下，循环执行某段程序，以处理需要重复处理的相同任务:

```python
# 目标价格
target_price = 15
# 当前价格
now_price = 1
# 涨幅
pctchange = 0.1
# 天数
day = 0
```

```python
# 超过目标价暂停
while now_price < target_price:
    # 每天涨10%
    now_price = round(now_price*(1+pctchange),2)
    # 计算天数
    day += 1
    print(day,now_price)
```

#### 2.1.2无限循环
如果条件判断语句永远为 true，循环将会无限的执行下去，如下实例：

```python
# 自动加一天
while day>=0:
    # 每天涨10%
    now_price = round(now_price*(1+pctchange),2)
    # 计算天数
    day += 1
    print(day,now_price)
```

#### 2.1.3循环使用 else 语句
else 中的语句会在循环正常执行完（即 while 不是通过 break 跳出而中断的）的情况下执行，for … else 也是一样。

```python
# 超过目标价暂停
while now_price < target_price:
    # 每天涨10%
    now_price = round(now_price*(1+pctchange),2)
    # 计算天数
    day += 1
    print(day,now_price)
else: 
    print("已达到盈利目标")
```

### 2.2for循环
for 循环语句允许我们执行一个语句或语句组多次（ 重复执行语句），称之为“遍历”或“迭代”。  
#### 2.2.1while语句执行过程
for循环的语法格式如下：

    for x in 某个对象:  
        循环执行的语句

```python
# 收盘价列表
closeprice = [23.45,27.45,78.45,100.23,998.89,2.34,5.44]
```

```python
# 通过循环快速计算出涨跌停
for x in closeprice:
    limitprice = round(x*1.1,2)
    print(limitprice)
```

#### 2.2.2列表内嵌for循环
可将多行简写成一句

```python
[round(price*1.1,2) for price in closeprice]
```

#### 2.2.3由序列的索引迭代

```python
# 也可以通过序列的索引进行迭代
for x in range(len(closeprice)):
    limitprice= round(closeprice[x]*1.1,2)
    print(limitprice)
```

#### 2.2.4循环使用 else 语句
else 中的语句会在循环正常执行完（即 for 不是通过 break 跳出而中断的）的情况下执行

```python
# 通过循环快速计算出涨跌停
for x in closeprice:
    limitprice = round(x*1.1,2)
    print(limitprice)
else:
    print('完成')
```

>## 3.循环的嵌套

Python 语言允许在一个循环体里面嵌入另一个循环。

+ 你可以在循环体内嵌入其他的循环体，如在while循环中可以嵌入for循环， 同样，也可以在for循环中嵌入while循环。

```python
pricelist = (30,65,53)
for price in pricelist:
    money = 60000
    while price*100 <= money:
        print("buy")
        money = money - (price*100)
    print(money)
```

+ 循环和条件同样可以相互嵌套。

```python
# 股票列表
stock = ['同花顺','工商银行','五粮液']
RSI = [49,99,68]
# 两个列表快速生成字典
stock_rsi_dict = dict(zip(stock,RSI))
stock_rsi_dict
```

```python
# 循环
for s in stock_rsi_dict:
    r = stock_rsi_dict[s]
    # 判断<20
    if r<=20:
        # 输出
        print('股票:{},RSI:{},目前弱，不建议买入'.format(s,r))
        pass
    # 再次判断 20-50
    elif r > 20 and r < 50:
        # 输出
        print('股票:{},RSI:{},目前较弱，建议观望'.format(s,r))
        pass
    # 再次判断
    elif r == 50:
        # 输出
        print('股票:{},RSI:{},目前方向不明确，建议观望'.format(s,r))
        pass
    # 再次50-70
    elif r > 50 and r < 70:
        #输出
        print('股票:{},RSI:{},目前较好，建议买一点'.format(s,r))
        pass
    # 再次判断 >70
    elif r>=70:
        # 输出
        print('股票:{},RSI:{},目前好，建议买入'.format(s,r))
        pass
```

### 4.1break语句
+ break语句用来终止循环语句，即循环条件没有False条件或者序列还没被完全递归完，也会停止执行循环语句。  
+ 如果正在使用嵌套循环，break语句将停止执行最深层的循环，并开始执行下一行代码。

该语句的格式为：<br><br>

    while 条件为真:
        执行语句...
        if中断语句条件 : 
            break

```python
# 超过目标价暂停
while now_price < target_price:
    # 每天涨10%
    now_price = round(now_price*(1+pctchange),2)
    # 计算天数
    day += 1
    # 天数超过6天，则停止
    if day > 6:
        break
    print(day,now_price)
```

### 4.2continue语句
+ continue 语句中断本次循环，而break中断整个循环。  
+ continue 语句用来告诉Python跳过当前循环的剩余语句，然后继续进行下一轮循环。

```python
# 目标价格
target_price = 15
# 当前价格
now_price = 1
# 涨幅
pctchange = 0.1
# 天数
day = 0


while now_price < target_price:
    now_price = round(now_price*(1+0.2),2)      # 每天涨20%
    day += 1
    if day == 6:
        continue
    print(day,now_price)
```
